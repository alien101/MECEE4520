```python
def get_loss(X, y, b0, b1, a, iter):

  betas = np.array([[b0], [b1]])
  for i in range(iter):
    #y_hat = X @np.array([[new_b0], [new_b1]])
    #new_b0 = new_b0 + a*(np.sum(y - y_hat)/100)
    #new_b1 = new_b1 + a*(np.sum((y - y_hat)*X[:,1]))/100
    grad = -1/100 *X.T @(y - X@ betas)
    betas = betas - a * grad
    #assert grad.shape == betas.shape
    #print(grad.shape)
  
  return betas
```

```python
def get_loss_square(X, y, b0, b1, b2, a, iter):

  for i in range(iter):
    betas = np.array([[b0], [b1], [b2]])
    #print(betas.shape)
    y_hat = X @ betas

    grad = -1. * (X.T @ (y - 1/(1 + np.exp(-y_hat))))

    betas = betas - a*grad

  return betas
  
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```
